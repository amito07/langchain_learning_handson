```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001"),
    persist_directory="./chroma_db",
    collection_name="tech_docs"
 )
```

```python
# Add documents synchronously so inserts are executed immediately
ids = vector_store.add_documents(docs)
print(f"Inserted {len(ids)} documents")
print(f"Collection count: {vector_store._collection.count()}")
```

```python
# View stored documents and confirm what is in the collection
db_snapshot = vector_store.get(include=['metadatas', 'documents'])
print(f"Returned documents: {len(db_snapshot.get('documents', []))}")
print(f"Stored IDs: {db_snapshot.get('ids', [])}")
db_snapshot
```

```python
# Perform a similarity search to find relevant documents based on a query
vector_store.similarity_search("What are the benefits of edge computing?", k=2)
```
